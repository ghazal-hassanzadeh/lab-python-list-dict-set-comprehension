````python
#1.
products = ["t-shirt", "mug", "hat", "book", "keychain"]
print(f"list of the products: {products}")
````

````python
def initialize_inventory(products):
    inventory = {
        product: int(input(f"Enter the quantity of {product}s available: "))
        for product in products
    }
    return inventory
inventory = initialize_inventory(products)
print(inventory)
````

````python
#2.
def get_customer_orders():
    number_of_orders = int(input("Enter the number of customer orders: "))
    
    customer_orders = {
        input("Please enter the product's name that you want to order: ")
        for i in range(number_of_orders)
    }
    
    return customer_orders

customer_orders = get_customer_orders()
print(customer_orders)
````

````python
#2.

def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1


    inventory = {prod: quant for prod, quant in inventory.items() if quant > 0}
    return inventory

inventory = update_inventory(customer_orders, inventory)
print(inventory)
````

````python
#2. 
def calculate_order_statistics(customer_orders, products):
    total = len(customer_orders)
    percentage = (total / len(products)) * 100
    return total, percentage

order_statistics = calculate_order_statistics(customer_orders, products)
print(order_statistics)
````

````python
#2.
def print_order_statistics(order_statistics):
    total, percentage = order_statistics
    print("Order Statistics:")
    print(f"Total unique products ordered: {total}")
    print(f"Percentage of catalog ordered: {percentage:.2f}%")

print_order_statistics(order_statistics)
````

````python
#3.
def calculate_total_price(customer_orders):
    prices = {
        product: float(input(f"Enter the price of {product}: "))
        for product in customer_orders
    }
    total_price = sum(prices.values())
    return total_price
````

````python
#4.

def update_inventory(customer_orders, inventory):
    for product in customer_orders:
        if product in inventory:
            inventory[product] -= 1


    inventory = {prod: quant for prod, quant in inventory.items() if quant > 0}
    return inventory

inventory = update_inventory(customer_orders, inventory)
print(inventory)
````
